<p>La finalidad de este documento es explicar y ejemplificar los usos de la función "<a href="https://www.tensorflow.org/api_docs/python/tf/random_normal">random_normal</a>", provista en la librería de <a href="https://www.tensorflow.org/">TensorFlow</a>, para Python.</p>
<p>Dicha función es utilizada para generar un tensor con valores aleatorios, que siguen una <a href="https://es.wikipedia.org/wiki/Distribuci%C3%B3n_normal">distribución normal</a>, con valores de media y desviación estandar conocidos.</p>

<p>La distribución de los valores generados, se podrá ver con mayor detalle en los ejemplos del final del documento.</p>

## Argumentos
<code><p style="font-size:20px">tf.random_normal(shape, mean=0.0, stddev=1.0, dtype=tf.float32, seed=None, name=None)</p></code>

Los argumentos de la función son:

<ul><li><p><b><code style="font-size:18px">shape</code></b>: Las dimensiones del tensor de salida. Se puede pasar a la función como un tensor 1-D de números entros, o como una lista de Python de números enteros. Este es el único argumento requerido de la función, los demás son opcionales.</p></li>
<p></p>
<li><p><b><code style="font-size:18px">mean</code></b> [ Valor predeterminado: <font color="blue"><b>mean = 0.0</b></font> ] : La media de la distribución normal, que seguirán los números generados. Se puede pasar a la función como un tensor 0-D, o un valor de Python, de tipo <b><code style="font-size:15px">dtype</code></b>.</p>
<p></p>
<li><p><b><code style="font-size:18px">stddev</code></b> [ Valor predeterminado: <font color="blue"><b>stddev = 1.0</b></font> ] : Es la desviación estándar de la distribución normal, que seguirán los números generados. Se puede pasar a la función como un tensor 0-D, o un valor de Python, de tipo <b><code style="font-size:15px">dtype</code></b>.</p></li>
<p></p>
<li><p><b><code style="font-size:18px">dtype</code></b> [ Valor predeterminado: <font color="blue"><b>dtype = tf.float32</b></font> ] : El tipo de valores que quiero generar. Puede ser <b><code style="font-size:15px">float32</code></b>, <b><code style="font-size:15px">float64</code></b>, <b><code style="font-size:15px">int32</code></b>, o <b><code style="font-size:15px">int64</code></b>.</p></li>
<p></p>
<li><p><b><code style="font-size:18px">seed</code></b> [ Valor predeterminado: <font color="blue"><b>seed = None</b></font> ] : Un número entero de Python. Es el generador de numeros aleatorios. Se utiliza para que sucesivas ejecuciones del programa generen siempre los mismos valores aleatorios, en el mismo orden. Es util cuando se esta depurando el programa, pero no se recomienda utilizarlo cuando el programa esta terminado.</p></li>
<p></p>
<li><p><b><code style="font-size:18px">name</code></b> [ Valor predeterminado: <font color="blue"><b>name = None</b></font> ] : El nombre de la operación. Lo utiliza TensorFlow para identificar distintas operaciones dentro del mismo programa. Se pasa a la función como una cadena de texto de Python.</p></li>

## Ejemplos de uso

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

El ejemplo de uso más básico es el de utilizar la funcion con todos los parametros con valores predeterminados. De esta forma estamos creando un tensor de números aleatorios, cuyos valores siguen una distribución normal con media  <b><code style="font-size:15px">0.0</code></b>, desviación estandar <b><code style="font-size:15px">1.0</code></b>, y son de tipo <b><code style="font-size:15px">float32</code></b>. Se ejemplificará además, el uso del atributo <b><code style="font-size:15px">name</code></b>.

```python
# Creación de un tensor de 2 x 5, con una lista de Python de números enteros
tensor_normal_1 = tf.random_normal([2, 5])

# Creación de un tensor de 5 x 3, con un tensor de números enteros
tensor = tf.constant([5, 3])
tensor_normal_2 = tf.random_normal(tensor, name="nombre_ejemplo")

# Visualización de los tensores creados
print("Tensor 1: {}".format(tensor_normal_1))
print("Tensor 2: {}".format(tensor_normal_2))
```

<p>Como se puede ver, los objetos creados con la función <b><code style="font-size:15px">tf.random_normal</code></b>, son tensores de tipo <b><code style="font-size:15px">float32</code></b>, y dimensiones (2, 5) y (5, 3), respectivamente. Además se puede ver el nombre de cada tensor. Cuando no se especifico el atributo <b><code style="font-size:15px">name</code></b>, el tensor obtuvo su nombre de la función que lo creó. Cuando se especifica un nombre, el tensor lo adquiere como identidificación.</p>
<p>Los tensores actualmente no contienen ningun valor, para generarlos, se deben correr los mismos en una Session de TensorFlow.</p>

```python
# Generación de los valores aleatorios
with tf.Session() as sess:
    tensor_normal_1 = sess.run(tensor_normal_1)
    tensor_normal_2 = sess.run(tensor_normal_2)

# Impresión de los resultados
print("Tensor con valores aleatorios N° 1: (shape={})".format(tensor_normal_1.shape))
print(tensor_normal_1)
print()
print("Tensor con valores aleatorios N° 2: (shape={})".format(tensor_normal_2.shape))
print(tensor_normal_2)
```

Como se puede observar, al no haber especificado valores para la media y la desviación estandar, la función utilizó los valores predeterminados de <b><code style="font-size:15px">mean</code></b> y <b><code style="font-size:15px">stddev</code></b>. Se puede ver cómo todos los valores generados están ubicados a ambos lados de la media. Sin embargo para visualizar efectivamente la distribición uniforme que siguen los mismos, es necesario generar más valores, para que la cantidad sea significativa.

### Observando la distribución de los valores
Para comprender mejor la distribucion normal que siguen los valores generados, se graficarán en un <a href="https://es.wikipedia.org/wiki/Histograma">histograma</a> de 10 clases, los valores generados en 2 tensores con 100 y 1000 números, respectivamente. Además se especificarán los atributos <b><code style="font-size:15px">media</code></b> y <b><code style="font-size:15px">stddev</code></b>, en uno de los tensores, para ejemplificar su uso.

```python
# Creación de ambos tensores
tensor_100 = tf.random_normal([100])
tensor_1000 = tf.random_normal([1000], mean=15, stddev=4)

# Generación de los valores aleatorios
with tf.Session() as sess:
    tensor_100 = sess.run(tensor_100)
    tensor_1000 = sess.run(tensor_1000)

# Importación de función de ayuda
from histogram_plots import histograms_x2
histograms_x2(tensor_100, tensor_1000)
```

<p>Como se puede ver en los gráficos, los valores generados, siguen una distribución normal, con centro en <code style="font-size:15px"><b>mean</b></code>. Al aumentar la cantidad de valores generados nos acercamos todavía más a la distribución normal teórica.</p>
<p>Las figuras muestran además, que el promedio de los valores aleatorios creados, está muy cerca del valor de la media. Esto es otra muestra del comportamiento normal de la población de números generados.</p>
